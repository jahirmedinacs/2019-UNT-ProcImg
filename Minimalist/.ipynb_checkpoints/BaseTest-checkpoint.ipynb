```python
plt.imshow(im[: , :, :3])
```

```python
[r, g, b] = [im[: , :, ii] for ii in range(3)]
```

```python
def hex_rgb(r,g,b):
    output = r * np.power(2,16)
    output += g * np.power(2,8)
    output += b
    return [output, output.astype(dtype='<U8')]
```

```python
[ref_int, _] = hex_rgb(r,g,b)
```

```python
tt = ref_int.astype(np.float)
```

```python
tt = (tt - tt.min())/(tt.max() - tt.min())
tt *= 255.0
tt = tt.astype(np.uint8)
```

```python
plt.imshow(tt,cmap='gray', vmin=0, vmax=255)
```

```python
def special_hist(data):
    __rr = np.unique(data)
    __rr = __rr.tolist() + [__rr.max() + 1]
    return np.histogram(tt, bins=__rr)
```

```python
hist_8 = special_hist(tt)
plt.plot(hist_8[1][1:], hist_8[0], '.b')
plt.plot(hist_8[1][1:], hist_8[0], '--c')
```

```python
def __reshape(x):
    return x.reshape(x.shape[0])
```

```python

__ii = [__reshape(hist_8[0]).tolist(), __reshape(hist_8[1]).tolist()[:-1]]
__ii = np.matrix(__ii)
display(__ii.shape)
__ii = pd.DataFrame(__ii.T, columns=["Hist", "Bin"])
__ii.sort_values(by=["Hist"], ascending=True, inplace=True)

__ii.describe()
```

```python
__jj = __ii.Hist.values / __ii.Hist.sum()
```

```python
# Criterio de 80 ~ 20
cut_point = 0.2
__ll = __jj.cumsum() > cut_point
plt.plot(__jj.cumsum())
plt.plot(__ll)
```

```python
__ii["P"] = __jj
__ii["crit"] = __ll
```

```python
centers = __ii[__ii.crit].Bin.values

ratio = tt.shape[0]/tt.shape[1]
h = 20
w = h * ratio
```

```python
plt.figure(figsize=(h,w))
for val in centers:
    print("GRAY VAL \t {:d}".format(val))
    plt.clf()
    plt.close()
    plt.imshow((tt == val).astype(np.int))
    plt.show()
```

```python
plt.clf()
plt.close()
```

```python
compose = np.zeros(tt.shape, dtype=np.uint8)
for val in centers:
    compose += (tt == val).astype(np.uint8)

display(np.unique(compose))

plt.figure(figsize=(h,w))
plt.imshow((compose + 1) % 2)
plt.show()
plt.close()
plt.clf()

plt.figure(figsize=(h,w))
plt.imshow(compose + 1)
plt.show()
plt.close()
plt.clf()
```
